## 2. Try to improve the contrast of each image by applying histogram equalization 

Plot the histograms of all images after HE and comment on whether HE has worked.

Let's start by creating a method that takes image in and presents the original image and the histogram and the equalized image and its histogram.

```python
def drawEqualizedHist(image):
    fig = plt.figure(figsize=(12,12))
    equ = cv.equalizeHist(image) # equalize image
    
    # Original image
    ax1 = fig.add_subplot(2,2,1)
    ax1.imshow(image, cmap='gray')
    ax1.set_title('Original Image')
    ax1.axis('off')
    
    # Equalized image
    ax2 = fig.add_subplot(2,2,2)
    ax2.imshow(equ, cmap='gray')
    ax2.set_title('Equalized Image')
    ax2.axis('off')
    
    # Original histogram 
    ax3 = fig.add_subplot(2,2,3)
    ax3.hist(image.flatten(),256,[0,256], color = 'r')
    ax3.set_title('Original Histogram')
    
    # Equalized histogram
    ax4 = fig.add_subplot(2,2,4)
    ax4.hist(equ.flatten(),256,[0,256], color = 'r')
    ax4.set_title('Equalized Histogram')

    plt.show()
```
